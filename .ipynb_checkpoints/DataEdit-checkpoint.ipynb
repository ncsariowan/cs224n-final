```python
import pandas as pd
import numpy as np
import random
import math
```

```python
case = "train"
```

```python
path = "./data/nsynth-" + case + "/examples.json"
df = pd.read_json(path)
df = df.transpose()
```

```python
import json

with open('instruments.json') as f:
    instruments = json.load(f)
    print(instruments)

with open('note_qualities.json') as f:
    note_qualities = json.load(f)
    print(note_qualities)
```

```python
df["location"] = "/data/nsynth-"+case+"/audio/" + df["note_str"] + ".wav"
df["dataset"] = "nsynth"
```

```python
vowels = ["a","e","i","o","u"]
notes = ["c", "c#", "d", "d#", "e", "f", "f#", "g", "g#", "a", "a#", "b"]
notesalt = ["c", "db", "d", "eb", "e", "f", "gb", "g", "ab", "a", "bb", "b"]

def makeArticle(word):

    return ("an" if (word[0].lower() in vowels) else "a")

def makeNote(number):
    letter = notes[number % 12]
    number = math.floor(number / 12) - 3
    return str(letter) + (str(number) if random.random() > 0.5 else "")

def addDescriptors(desArr):
    s = ""
    for d in desArr:
        s += ", " + random.choice(note_qualities[d].get("synonyms"))

    if len(s) > 0:
        s = s[2:]

    return s

#wip
def makeCaption(data):
    i = data["instrument_family_str"].map(instruments)
    i = pd.Series([random.choice(x.get('names')) for x in i], index=data.index)
    ai = pd.Series([makeArticle(x) for x in data["instrument_source_str"]], index=data.index)
    n = pd.Series([makeNote(x) for x in data["pitch"]], index=data.index)
    ni = pd.Series([makeArticle(x) for x in n], index=data.index)
    d = pd.Series([addDescriptors(x) for x in data["qualities_str"]], index=data.index)
    
    s = d + " " + data["instrument_source_str"] + " " + i
    s = (s + " playing " + ni + " " + n)
    si = pd.Series([makeArticle(x.strip()) for x in s], index=data.index)
    s = si + " " + s

    return s

# makeCaption(df.iloc[0:10])
df["captions"] = makeCaption(df)
```

```python
df[["location", "captions"]].transpose().to_json('nsynth-' + case+'.json')
```

```python
def process(case):
    path = "./tango/data/nsynth-" + case + "/examples.json"
    df = pd.read_json(path)
    df = df.transpose()

    df["location"] = "./data/nsynth-"+case+"/audio/" + df["note_str"] + ".wav"
    df["dataset"] = "nsynth"
    df["captions"] = makeCaption(df)
    
    # for jdict in df[["location", "captions"]].transpose().to_dict(orient='records'):
    #     print(jdict)
        
    df[["location", "captions"]].to_json('nsynth-' + case+'.json', orient="records",
           lines=True)

process("test")
process("valid")
process("train")
```
